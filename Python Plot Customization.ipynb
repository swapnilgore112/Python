```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y1=[2,3,5,7,11]
y2=[1,4,6,8,10]
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y1, label="Prime Number", color='green',linestyle='--',marker='o')
plt.plot(x,y2,label='Even Number',color='red',linestyle='-',marker='x')
plt.title("Customized plot")
plt.grid(True,which='both',linestyle='--',linewidth=0.5)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```
